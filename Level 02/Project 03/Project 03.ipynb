# 3. Model Implementation :

## I. Scaling the factors :
### Only Time and Amount need scaling because others are already PCA scaled.

```python
scaler = StandardScaler()

data_copy_1['Amount_Scaled'] = scaler.fit_transform(data_copy_1['Amount'].values.reshape(-1,1))
data_copy_1['Time_Scaled'] = scaler.fit_transform(data_copy_1['Time'].values.reshape(-1,1))
data_copy_1.drop(['Amount','Time'],axis = 1 , inplace=True)
```

```python
data_copy_1
```

## II. Handle Class Imbalance :

```python
x = data_copy_1.drop('Class',axis = 1)
y = data_copy_1['Class']

sm = SMOTE()

x_res , y_res = sm.fit_resample(x,y)
```

## III. Train-Test-Split :

```python
x_train , x_test , y_train , y_test = train_test_split(x_res , y_res , random_state= 42 , test_size= 0.2 )
```

# 4. Training Model :

## I. Random Forest :

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

## 5. Making Predictions :

```python
y_pred = rf.predict(x_test)
```

### The trained model takes test data (X_test).

### It predicts whether each transaction is:

### 0 → Normal transaction

### 1 → Fraud transaction

### The predicted results are stored in y_pred.

## 6. Model Evaluation :

### I. Confusion Matrix :

```python
confusion_matrix(y_test, y_pred)
```

### II. Classification Report :

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
