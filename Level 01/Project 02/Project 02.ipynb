## 2. Data Exploration :

### i. Boxplot for the total amount spend :

```python
# we use boxplot inorder to find any possible outliers :

plt.figure(figsize=(6,4))
sns.boxplot(data = cust_data,y = 'MntTotal')
plt.title('Box Plot for MntTotal')
plt.ylabel("Mnt Total")
plt.show()
```

#### The Box Plot noticed some outliers.

```python
Q1 = cust_data['MntTotal'].quantile(0.25)
Q3 = cust_data['MntTotal'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = cust_data[(cust_data['MntTotal']<lower_bound) | (cust_data['MntTotal']>upper_bound)]
outliers.head()
```

### ii. Removing Outliers :

```python
cust_data_clean = cust_data[(cust_data['MntTotal'] > lower_bound) & (cust_data['MntTotal'] < upper_bound)]
cust_data_clean.describe()
```

### After Outlier Handling :

* Extreme MntTotal values were removed from the cleaned dataset

* Dataset size was slightly reduced (only a small number of rows removed)

* Key improvements:

    i.    More balanced data distribution
    
    ii.   Reduced skewness
    
    iii.  Mean and median values became closer

## 4. Feature Engineering :

### i. Marital :

#### The data frame contains 5 columns to reflect marital status. We are going to create a new column 'marital' with values: Divorced, Married, Single, Together, Widow. This column will allow us to draw some additional plots.

```python
cust_data_clean = cust_data_clean.copy()

cust_data_clean['Marital'] = 'Unknown'

cust_data_clean.loc[cust_data_clean['marital_Divorced'] == 1, 'Marital'] = 'Divorced'
cust_data_clean.loc[cust_data_clean['marital_Married'] == 1,  'Marital'] = 'Married'
cust_data_clean.loc[cust_data_clean['marital_Single'] == 1,   'Marital'] = 'Single'
cust_data_clean.loc[cust_data_clean['marital_Together'] == 1, 'Marital'] = 'Together'
cust_data_clean.loc[cust_data_clean['marital_Widow'] == 1,    'Marital'] = 'Widow'
```

### Barplot :

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Marital', y='MntTotal', data=cust_data_clean, palette='viridis')
plt.title('MntTotal by marital status')
plt.xlabel('Marital status')
plt.ylabel('MntTotal')
plt.show()
```

### Conclusion – 'MntTotal' by Marital Status :

* Widowed customers have the highest average total spending among all marital groups.

* Divorced customers also show relatively high spending compared to others.

* Single and Together customers have similar spending patterns.

* Married customers show slightly lower average spending compared to other groups.

* The differences are moderate, but marital status appears to have some influence on spending behavior.

### ii. In_relationship :

#### There are 3 features that reflect if a person is single (Single, Divorced, Widow) and 2 features if a person is in relationship (Together, Married). We will add an additional feature 'In_relationship'. This feature will equal 1 if a customer's marital status is 'Married' or 'Together' and 0 in all other cases.

```python
def get_relationship(row):
    if row['marital_Married'] ==1:
        return 1
    elif row['marital_Together'] == 1:
        return 1
    else:
        return 0
cust_data_clean['In_relationship'] = cust_data_clean.apply(get_relationship, axis=1)
cust_data_clean.head()    
```

## 5. K-means Clustering :

### i. Standardising data :

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols_for_clustering = ['Income','MntTotal','In_relationship']
data_scaled = cust_data_clean.copy()
data_scaled[cols_for_clustering] = scaler.fit_transform(cust_data_clean[cols_for_clustering])
data_scaled[cols_for_clustering].describe()
```

#### The mean value for all colums is almost zero and the standard deviation is almost 1. All the data points were replaced by their z-scores.

### ii. Principal Component Analysis :(PCA)

```python
from sklearn import decomposition

pca = decomposition.PCA(n_components = 2)
pca_res = pca.fit_transform(data_scaled[cols_for_clustering])
data_scaled['pc1'] = pca_res[:,0]
data_scaled['pc2'] = pca_res[:,1]
```

### iii. Cluster method :

```python
X = data_scaled[cols_for_clustering]
inertia_list = []
for K in range(2,10):
    inertia = KMeans(n_clusters=K, random_state=7).fit(X).inertia_
    inertia_list.append(inertia)
```

```python
plt.figure(figsize=[7,5])
plt.plot(range(2,10), inertia_list, color=(54 / 255, 113 / 255, 130 / 255))
plt.title("Inertia vs. Number of Clusters")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.show()
```

### iv. Silhouette score analysis :

```python
from sklearn.metrics import silhouette_score
silhouette_list = []
for K in range(2,10):
    model = KMeans(n_clusters = K, random_state=7)
    clusters = model.fit_predict(X)
    s_avg = silhouette_score(X, clusters)
    silhouette_list.append(s_avg)

plt.figure(figsize=[7,5])
plt.plot(range(2,10), silhouette_list, color=(54 / 255, 113 / 255, 130 / 255))
plt.title("Silhouette Score vs. Number of Clusters")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.show()
```

#### The highest silhouette score is for 4 clusters.
